```python
import pandas as pd
url = "https://raw.githubusercontent.com/NathaliaMinoque/datasets/refs/heads/main/jabodetabek_house_price.csv"
df = pd.read_csv(url)
df.head()
```

```python
# Calculate null values and percentage
null_counts = df.isnull().sum()
null_percentages = (null_counts / len(df)) * 100

# Create a DataFrame for columns with null values
null_df = pd.DataFrame({'Null Count': null_counts, 'Null Percentage': null_percentages})

# Filter for columns with null values
null_df = null_df[null_df['Null Count'] > 0]


# Format the percentage with two decimal places and % symbol
null_df['Null Percentage'] = null_df['Null Percentage'].map('{:.2f}%'.format)


null_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create the horizontal bar plot
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
sns.barplot(x='Null Count', y=null_df.index, data=null_df, orient='h', palette='rainbow')
plt.xlabel('Number of Null Values', fontsize=12)
plt.ylabel('Columns', fontsize=12)
plt.title('Null Values per Column', fontsize=14)

# Add labels to the bars
for i, v in enumerate(null_df['Null Count']):
    plt.text(v + 1, i, str(v), color='black', va='center', fontsize=10)

plt.tight_layout()
plt.show()
```

KEY FINDINGS
*  The dataset has 3553 entries and 27 features, with both numerical and categorical data. This dataset is large enough for reliable analysis and it allows us to study housing from their physical characteristics and qualitative aspects.

*   Some columns have missing values, especially for the columns year_built, building_age, certificate, furnishing, and property_conditions. Missing data can reduce the accuracy of analysis in which the results may be misleading, so we need to do imputation or removal of data to avoid bias and improve data quality.

* Land and building size shows extremely different variation where most houses are small to medium sized, but a there are a few very large properties. This means that some houses are much bigger than most others, and they can make the average size look misleading. That is why the median is a better choice to describe the typical house size, because it is not affected much by very large properties.

*   House prices are highly skewed: with average price of Rp 4.19B, but the median is only Rp 1.5B, and outliers reach up to Rp 580B. This shows that most houses are not actually that expensive, and only a few outlier houses raise the average a lot. Looking at the median is more useful to know the real market price. This also means that there is a very wide gap between normal houses and luxury houses in Jabodetabek.


*   Bedrooms and bathrooms for every houses mostly range from 2-4, but some housed have extreme values of 99 which data entry errors. If we keep them, they will affect our analysis and make the results unreliable. Cleaning or removing this wrong data is important to make sure we only analyze realistic values.

*   Features like maid rooms, maid bathrooms, and carports are often 0, meaning not all house include them. This tells us that not every house has extra rooms or facilities.


*   Year built values suggest most houses are new (median 2021), but errors exist (e.g., 1870 and 2052). If not corrected, these errors can give misleading insights about how building age affects house prices. Cleaning this data is important before further analysis.

EXPLORATORY DATA ANALYSIS (EDA)

```python
# Relationship between building_age_category and price_per_m2

plt.figure(figsize=(8,6))
sns.boxplot(x='building_age_category', y='price_per_m2', data=df, palette='Set2')
plt.ylim(0, df['price_per_m2'].quantile(0.95))  # limit extreme outliers for clarity
plt.title("Price per mÂ² by Building Age Category")
plt.xlabel("Building Age Category")
plt.ylabel("Price per mÂ² (Rp)")
plt.xticks(rotation=20)
plt.show()
```

The boxplot shows that houses aged 5â€“15 years usually have the highest price per mÂ² compared to other age groups. Houses in this range may be more attractive because they are still in good condition but not as expensive as brand-new ones. In contrast, older houses (15â€“30 years) are usually cheaper per mÂ², likely because of depreciation or the need for renovation. Whatâ€™s interesting is that new houses (<5 years) do not always have the highest price per mÂ², as their median price is almost the same as older properties. In addition, very old houses (30+ years) sometimes have similar price levels to moderately new houses, which shows that building age is not the only factor that affects price.

This analysis is important because it shows that building age does not directly determine the property value. It shows us that maybe other factors such as location and demand of property may influence the house price. For buyers, this insight suggests that a moderately new property could be a better investment than paying extra for a brand-new house, depending on its location. or investors and developers, this means that focusing only on building age could be misleading, they should consider multiple aspects of a property rather than focusing only on age when judging its market value.

```python
# Histogram showing distribution of house prices

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.figure(figsize=(8,5))
sns.histplot(np.log1p(df['price_in_rp']), bins=50, kde=True)
plt.title("Distribution of House Prices (log scale)")
plt.xlabel("Log(Price in Rp)")
plt.ylabel("Count")
plt.show()
```

The histogram shows that house prices in Jabodetabek are not evenly distributed and are strongly skewed. Most houses are priced in the middle range, while only a small number of houses fall into the very high-price category. After applying the log scale, the distribution looks closer to normal, which makes it easier to see the overall price pattern.

This analysis is important because it helps us understand the general market trend and detect extreme outliers. By using log transformation, we can analyze prices more clearly without being dominated by very expensive houses. This is useful for both researchers and buyers to get a fairer picture of the housing market. It shows that setting the right price is important to stay competitive, especially since most houses are in the middle price range.

```python
# Boxplot of price by city
plt.figure(figsize=(10,6))
sns.boxplot(x='city', y=np.log1p(df['price_in_rp']), data=df)
plt.title("House Prices by City (log scale)")
plt.ylabel("Log(Price in Rp)")
plt.xticks(rotation=45)
plt.show()
```

The chart also shows many outliers in every city, meaning that even in cheaper cities, there are still some luxury houses that cost much more than the average. This tells us that the housing market in Jabodetabek is very diverse, and prices are not only based on the city but also the specific neighborhood.

This analysis is important because it shows the role of location in determining house prices. It shows that the same city can have both cheap and very expensive houses, depending on the neighborhood. This information can be use to compare affordability of house price between cities. For buyers, this information helps to compare the affordability of housing between cities and decide where they can get the best value.
